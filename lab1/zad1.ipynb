```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = pow(10, 7)
m = np.random.uniform(0.1, 0.9)
tab = [m for _ in range(N)]
```

Wylosowana liczba i właściwy wynik obliczeń

```python
print(m, N*m)
```

```python
def cumulative_sum():
    acc = 0
    for x in tab:
        acc += x
    return acc
```

Błąd bezwzględny i względny

```python
res = cumulative_sum()
err = abs(res - N*m)
print(err)
print(err / N)
```

Wzrost błędu względnego(krok co 25000 łącznie 400)

```python
acc2 = 0
error = []
for i in range(N):
    acc2 += tab[i]
    if i % 25000 == 0:
        t = (i + 1)*m
        error.append(abs(acc2 - t)/t)

plt.plot(error)
plt.show()
```

```python
def recursive_sum(l = 0, r = N - 1):
    if l == r:
        return tab[l]
    s = (l + r) // 2
    return recursive_sum(l, s) + recursive_sum(s + 1, r)
```

Błąd bezwględny i względny metodą rekurencyjną

```python
res2 = recursive_sum()
err2 = abs(res2 - N*m)
print(err2)
print(err2 / N)
```

Porównanie czasu działania

```python
from time import time
t1 = time()
cumulative_sum()
t1 = time() - t1
print("Metoda iteracyjna: ", t1)
t1 = time()
recursive_sum()
t1 = time() - t1
print("Metoda rekurencyjna: ", t1)
```
